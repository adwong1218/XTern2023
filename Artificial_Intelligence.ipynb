```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
# Import Data into dataframe
df = pd.read_csv("~/Xtern2023/XTern 2024 Artificial Intelegence Data Set - Xtern_TrainData.csv")
df.head()
```

```python
# Listing the Unique Entries for Each Column
for column in df.columns:
    print(f"Unique values for {column}: {df[column].unique()}\n")
    print(f"Number of Unique values for {column}: {len(df[column].unique())}\n\n")
```

```python
# Info of Dataset
df.info()

# Print number of missing values
print("\nNumber of Missing Values:")
print(df.isnull().sum())

# Print unique values
print("\nUnique Values:")
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Column '{column}': {unique_values}")
    print(f"Number of Unique Values in '{column}': {len(unique_values)}\n")
```

```python
# Graphing Year Vs Order

# Create a scatter plot
plt.scatter(df['Year'], df['Order'], marker='o', color='blue', label='Ordered Items')

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('Order')
plt.title('Scatter Plot of Year vs Order')

# Rotate the y-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Graphing Major Vs Order

# Create a scatter plot
plt.scatter(df['Major'], df['Order'], marker='o', color='blue', label='Ordered Items')

# Add labels and a title
plt.xlabel('Major')
plt.ylabel('Order')
plt.title('Scatter Plot of Major vs Order')

# Rotate the y-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
# Graphing Major Vs Order

# Create a scatter plot
plt.scatter(df['Time'], df['Order'], marker='o', color='blue', label='Ordered Items')

# Add labels and a title
plt.xlabel('Time')
plt.ylabel('Order')
plt.title('Scatter Plot of Time vs Order')

# Rotate the y-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Converting Columns Into Integers

# Creating instance of label converter
lab = LabelEncoder()

# Encoding labels for year
df_encoded = df
df_encoded['Year'] = lab.fit_transform(df_encoded['Year'])
df_encoded['Major'] = lab.fit_transform(df_encoded['Major'])
df_encoded['University'] = lab.fit_transform(df_encoded['University'])
df_encoded['Order'] = lab.fit_transform(df_encoded['Order'])

df_encoded.head()
```
